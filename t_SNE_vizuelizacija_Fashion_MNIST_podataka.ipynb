```python
# Importovanje potrebnih biblioteka
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.datasets import fetch_openml
from sklearn.preprocessing import StandardScaler

# Uitavanje Fashion-MNIST podataka
print("Ucitavanje Fashion-MNIST podataka...")
X, y = fetch_openml('Fashion-MNIST', version=1, return_X_y=True, as_frame=False)

# Izdvajam 5000 uzoraka, kako bi algoritam bre radio
X = X[:5000]
y = y[:5000]

# Izvrena normalizacija podataka
X = StandardScaler().fit_transform(X)

# Pokretanje t-SNE algoritma
print("Primjena t-SNE algoritma...")
tsne = TSNE(n_components=2, perplexity=30, n_iter=1000, random_state=42)
X_embedded = tsne.fit_transform(X)

# Prikaz vizuelizacije u 2D dimenziji
print("Crtanje rezultata...")
plt.figure(figsize=(10, 8))
scatter = plt.scatter(X_embedded[:, 0], X_embedded[:, 1], c=y.astype(int), cmap='tab10', s=10)
plt.colorbar(scatter, ticks=range(10), label='Kategorija')
plt.title('t-SNE vizuelizacija Fashion-MNIST podataka')
plt.xlabel('Prva komponenta')
plt.ylabel('Druga komponenta')
plt.grid(True)
plt.show()
```
